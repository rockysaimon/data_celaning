# cambiar nombres

```python
nombres_nuevos={
    "battery_power":"battery_mAh",
    "blue":"bluethoot",
    "clock_speed":"clock_speed",
    "dual_sim":"dual_sim",
    "fc":"front_camera_mp",
    "four_g":"4G",
    "int_memory":"memory_GB",
    "m_dep":"mobile_depth_cm",
    "mobile_wt":"mobile_weight_gr",
    "n_cores":"processor_cores",
    "pc":"primary_camera_mp",
    "px_height":"pixel_resolution_h",
    "px_width":"pixel_resolution_w",
    "ram":"ram_MB",
    "sc_h":"screen_height_cm",
    "sc_w":"screen_width_cm",
    "talk_time":"battery_time",
    "three_g":"3G",
    "touch_screen":"touch_screen",
    "wifi":"wifi",
    "price_range":"price_range"
}

df=data.rename(columns=nombres_nuevos)
df.head(10)
```

# especificar rango de precios

```python
rango_precio = {
    0: 'bajo',
    1: 'medio',
    2: 'alto',
    3: 'muy alto'
}

df["price_range"] = df["price_range"].replace(rango_precio)
df.head(10)
```

**Comprobar si hay datos duplicados**

```python
duplicados = df[df.duplicated()]
print("Filas duplicadas:")
duplicados
```

**eliminar datos duplicados**

```python
df = df.drop_duplicates()
print("DataFrame sin duplicados:")
df.info()
```

**comprobar**

```python
duplicados = df[df.duplicated()]
print("Filas duplicadas:")
duplicados
```

**eliminar filas con nulos**

```python
df = df.dropna()
print("DataFrame sin filas nulas:")
df.info()
```

# cambiar tipo de datos

```python
columnas_cambiar=["battery_mAh","processor_cores","pixel_resolution_h","pixel_resolution_w"]
df[columnas_cambiar]=df[columnas_cambiar].astype(int)
df.head()
```

# outliers

```python
import matplotlib.pyplot as plt

plt.scatter(df["price_range"], df["battery_mAh"])
plt.xlabel("price_range")
plt.ylabel("battery_mAh")
plt.title('Gráfico de dispersión')

```

```python
# Identificar outliers utilizando el método del rango intercuartílico (IQR)
Q1 = df["battery_mAh"].quantile(0.25)
Q3 = df["battery_mAh"].quantile(0.75)
IQR = Q3 - Q1

# Calcular límites del rango intercuartílico (IQR) con un factor personalizado y un límite inferior más alto que 0
factor_escala = 1.5  # Puedes ajustar este valor según tus necesidades
limite_inferior = max(800, Q1 - factor_escala * IQR)  # Establece el límite inferior como máximo entre 0 y el cálculo del IQR
limite_superior = Q3 + factor_escala * IQR

# Eliminar filas con outliers en las columnas seleccionadas
df = df[(df["battery_mAh"] >= limite_inferior) & (df["battery_mAh"] <= limite_superior)]

# Mostrar el DataFrame resultante sin outliers
df
```

```python
plt.scatter(df["price_range"], df["battery_mAh"])
plt.xlabel("price_range")
plt.ylabel("battery_mAh")
plt.title('Gráfico de dispersión')
```
