## Identificar Outliers

>En esta fase crítica de nuestro análisis de datos, nos hemos centrado en identificar y abordar los outliers en nuestras columnas de interés, tales como, battery_mAh, battery_time, ram_MB y processors_cores. Los outliers son valores extremos que se encuentran significativamente alejados del resto de los datos y pueden distorsionar nuestros análisis, por lo que es esencial identificarlos y manejarlos adecuadamente para mantener la integridad y la precisión de nuestros datos.

>Hemos utilizado una herramienta poderosa llamada el método del rango intercuartílico (IQR) para identificar los outliers en nuestra columna de interés. Calculamos el rango intercuartílico (IQR) y luego definimos límites superior e inferior basados en este rango y un factor de escala personalizado. Cualquier valor fuera de estos límites se considera un outlier y se elimina del conjunto de datos.

>A continuación, hemos vuelto a visualizar la distribución de nuestra columna de interés utilizando un boxplot para confirmar que los outliers han sido eliminados con éxito y que nuestros datos están ahora libres de distorsiones.

## **Outliers columna battery_mAh**



```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['battery_mAh'], vert=False)  # Utiliza vert=False para hacer el boxplot horizontal
plt.title('Boxplot de la columna específica')
plt.xlabel('Valores')
plt.ylabel('Columna específica')
plt.grid(True)
plt.show()
```

```python
# Identificar outliers utilizando el método del rango intercuartílico (IQR)
Q1 = df["battery_mAh"].quantile(0.25)
Q3 = df["battery_mAh"].quantile(0.75)
IQR = Q3 - Q1

# Calcular límites del rango intercuartílico (IQR) con un factor personalizado y un límite inferior más alto que 0
factor_escala = 1.5  # Puedes ajustar este valor según tus necesidades
limite_inferior = Q1 - factor_escala * IQR  #  Establece el límite inferior como máximo entre 0 y el cálculo del IQR
limite_superior = Q3 + factor_escala * IQR
print(limite_inferior,limite_superior)

# Eliminar filas con outliers en las columnas seleccionadas
df = df[(df["battery_mAh"] >= limite_inferior) & (df["battery_mAh"] <= limite_superior)]

# Mostrar el DataFrame resultante sin outliers
df.info()
```

# MODELO DE PREDICCIÓN

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```


### modelo de entrenamiento

>En esta etapa, creamos un modelo de redes neuronales profundas para predecir las categorías de precios de los teléfonos móviles. Preparamos los datos, codificamos las etiquetas de precios y dividimos el conjunto de datos en entrenamiento y prueba. Desarrollamos un modelo con 20 capas ocultas y lo entrenamos con datos de entrenamiento y validación. Este modelo nos permitirá hacer predicciones precisas sobre nuevas muestras de datos.

```python
le = LabelEncoder()
df['price_range_encoded'] = le.fit_transform(df['price_range'])

# Separar los datos en características (X) y etiquetas (y)
X = df.drop(['price_range', 'price_range_encoded'], axis=1)
y = df['price_range_encoded']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo con 25 capas
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=X_train.shape[1]))

# Añadir 25 capas ocultas con activación ReLU
for _ in range(25):
    model.add(tf.keras.layers.Dense(64, activation='relu'))

# Añadir capa de salida con activación softmax para clasificación multiclase
model.add(tf.keras.layers.Dense(4, activation='softmax'))

# Compilar el modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Resumen del modelo
model.summary()

# Entrenar el modelo
model.fit(X_train, y_train, epochs=21, batch_size=32, validation_data=(X_test, y_test))
```

>Después de entrenar nuestro modelo de redes neuronales profundas, verificamos su rendimiento utilizando el conjunto de datos de prueba. Obtenemos las predicciones del modelo y las comparamos con los resultados reales, decodificando las clases predichas y esperadas. Imprimimos cinco ejemplos para inspeccionar el rendimiento del modelo en casos individuales.

```python
categorias_string_inverso = {
    0: 'bajo',
    1: 'medio',
    2: 'alto',
    3: 'muy alto'
}
# Obtener las predicciones del modelo en el conjunto de datos de prueba
predicciones = model.predict(X_test)
# Convertir las probabilidades predichas a las clases predichas (índice de la clase con la probabilidad más alta)
predicciones_clases = np.argmax(predicciones, axis=1)

# Decodificar las clases predichas y los resultados esperados de números a categorías de string utilizando el diccionario categorias_string_inverso
predicciones_string = [categorias_string_inverso[indice] for indice in predicciones_clases]
resultados_esperados = [categorias_string_inverso[indice] for indice in y_test]

# Imprimir cinco ejemplos de predicciones y resultados esperados
for i in range(5):
    print("Ejemplo", i+1)
    print("Características:")
    print(X_test.iloc[i])  # Acceder a la fila del conjunto de datos de prueba
    print("Predicción:", predicciones_string[i])
    print("Resultado esperado:", resultados_esperados[i])
    print()
```
